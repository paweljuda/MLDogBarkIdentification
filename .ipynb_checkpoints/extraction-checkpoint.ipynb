```python
import librosa
import librosa.display
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import os
import IPython

T = 2.0    # seconds
sr = 22050 # sample rate
t = np.linspace(0, T, int(T*sr), endpoint=False) # time variable
x = 0.5*np.sin(2*np.pi*440*t)                # pure sine wave at 440 Hz
IPython.display.Audio(x, rate=sr)


for files in os.listdir("./AudioSet1"):
    
    y, sr = librosa.load("./AudioSet1/" + files)

    # Let's make and display a mel-scaled power (energy-squared) spectrogram
    S = librosa.feature.melspectrogram(y, sr=sr, n_mels=128)
    S = S[:,0:45]
    # Convert to log scale (dB). We'll use the peak power as reference.
    log_S = librosa.amplitude_to_db(S,ref=np.max)

    plt.figure(figsize=(10, 4))
    S_dB = librosa.power_to_db(S**2, ref=np.max)
    librosa.display.specshow(S_dB, x_axis='time',y_axis='mel', sr=sr,fmax=8000)
    plt.colorbar(format='%+2.0f dB')
    plt.title('Mel-frequency spectrogram for ' + files)
    print(S_dB.shape)
    IPython.display.Audio("./AudioSet1/Leo1.wav")
    plt.tight_layout()
    plt.show()
    
    z = librosa.feature.mfcc(y=y, sr=sr, S=S_dB)
    print(z.shape)

    break
```
